```python
def get_category(row):
    categories = [
        "Fossil fuel representatives", "Members of youth-focused organizations", "Students", "Other small and/or community-based NGOs", 
        "Government representatives", "Scientists", "Journalists/Media", "Representatives of international organizations", 
        "Representatives of intergovernmental organizations", "Other executives", "Other corporates", "Academics"
    ]
    
    # Combine relevant fields for categorization
    combined_text = " ".join(
        str(row[field]).lower() for field in ["Job_Title", "Affiliation", "Group_Type", "Division"] if pd.notna(row[field])
    )
    
    result = pipe(combined_text, categories)
    return result['labels'][0], result['scores'][0]
```

```python
import csv
import pandas as pd
from tqdm import tqdm

output_file = "categorized_results.csv"

# Write headers to the CSV file
with open(output_file, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Job Title", "Affiliation", "Group Type", "Division", "NGO", "IGO", "Category", "Confidence Score"])  # Ensure correct columns

chunk_size = 10_000
for i in range(0, len(df), chunk_size):
    df_chunk = sample.iloc[i:i + chunk_size].copy()  # Copy to avoid Pandas warning

    # Apply categorization function
    df_chunk[['Category', 'Confidence Score']] = df_chunk.progress_apply(
        lambda row: pd.Series(get_category(row)), axis=1
    )

    # Append results to CSV
    df_chunk.to_csv(output_file, mode='a', index=False, header=False)  

print("Processing complete. Results saved to", output_file)
```
